```python

from sklearn.svm import SVR
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

plt.style.use('fivethirtyeight')

# load the data
from google.colab import files
files.upload()
```

```python
# store data
df = pd.read_csv('KO.csv')
# show data
print(df)
```

```python
# show and store the last row of the data
actual_price = df.tail(1)
# show data 
print(actual_price)
```

```python
# get all of the data except last row 
df = df.head(len(df)-1)
# Show data 
print(df)
```

```python
# Create empty lists
days = list()
adj_close_prices = list()

# get only date and the adjusted close prices 
df_days = df.loc[:, 'Date']
df_adj_close = df.loc[:, 'Adj Close']
```

```python
# Create the independent data set (dates)
for day in df_days:
    days.append([int(day.split('-')[2])])
# Create the dependent dataset(adjusted)

for adj_close_price in df_adj_close:
    adj_close_prices.append(float(adj_close_price))
```

```python
# show days
print(days)

# show adjusted close prices
print(adj_close_prices)
```

```python
# Create 3 Models
lin_svr = SVR(kernel='linear', C=1000.0)
lin_svr.fit(days, adj_close_prices)

ploy_svr = SVR(kernel='poly', C=1000.0, degree=2)
ploy_svr.fit(days, adj_close_prices)

rbf_svr = SVR(kernel='rbf', C=1000.0, gamma=0.85)
rbf_svr.fit(days, adj_close_prices)
```

```python
# show the predicted price for the given data 
day = [[30]]
print('The RBF SVR predicted price:', rbf_svr.predict(day))
print('The Linear SVR predicted price:', lin_svr.predict(day))
print('The Ploynomial SVR predicted price:', ploy_svr.predict(day))
```
